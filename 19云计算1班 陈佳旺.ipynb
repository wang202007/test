# 图书馆销售数据

```python
import numpy as np
```

```python
sale_price =np.loadtxt('price.csv',delimiter=',',dtype=(float))
print('销售价格表：',sale_price)
```

```python
sale_price.sort()
print('排序后的价格为：',sale_price)
```

```python
print('去掉重复价格值的表为：',np.unique(sale_price))
```

```python
print('价格表总和为：',np.sum(sale_price))
```

```python
print('价格表均值为：',np.average(sale_price))
print('价格表均值为：',np.mean(sale_price))
```

```python
print('价格表标准差为：',np.std(sale_price))
```

```python
print('价格表方差为：',np.var(sale_price))
```

```python
print('价格表最小值为：',np.min(sale_price))
```

```python
print('价格表最大值为：',np.max(sale_price))
```

# 图片处理

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
```

```python
earth = mpimg.imread('earth.jpeg')
xmax = earth.shape[0]  #xmax表示行数
ymax = earth.shape[1]  #ymax表示列数
plt.imshow(earth)
```

```python
def shuffle_indices(size):
    arr = np.arange(size)
    np.random.shuffle(arr)
    return arr
```

```python
xindlices = shuffle_indices(xmax)
np.testing.assert_equal(len(xindlices),xmax)
yindlices = shuffle_indices(ymax)
np.testing.assert_equal(len(yindlices),ymax)
```

```python
plt.imshow(earth[np.ix_(xindlices,yindlices)])
plt.show()
```
